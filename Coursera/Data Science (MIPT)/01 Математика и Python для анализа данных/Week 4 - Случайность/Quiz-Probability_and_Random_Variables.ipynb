**3.** 1% женщин больны раком груди. У 80% женщин, больных раком груди, маммограмма верно выявляет наличие заболевания; кроме того, она даёт ложный положительный результат (то есть, неверно показывает наличие рака) для 9.6% здоровых женщин.

У какого процента женщин, маммограмма которых дала положительный результат, есть рак груди? Запишите ответ с точностью до одного знака после десятичной точки (знак процента не нужен).

Пусть событие A — наличие рака, B — положительный результат маммограммы. По формуле полной вероятности
$${\sf P}\left(B\right) = {\sf P} \left(B\left|A\right.\right) {\sf P} \left(A\right) + {\sf P} \left(B\left|\bar{A}\right.\right) {\sf P} \left(\bar{A}\right) = 0.8\cdot0.01+ 0.096\cdot0.99=0.10304.$$
Воспользуемся формулой Байеса:

$${\sf P} \left(A\left|B\right.\right) = \frac{{\sf P}\left(A\right) {\sf P} \left(B\left|A\right.\right)}{{\sf P}\left(B\right)} = \frac{0.01\cdot0.8}{0.10304}\approx 0.078.$$
Или 7.8 %

**4**.Какова вероятность того, что при независимом подбрасывании двух симметричных шестигранных кубиков хотя бы на одном из них выпадет больше трёх очков? Запишите точный ответ в виде десятичной дроби.

Необходимо использовать следующую формулу:
$$P(A+B)=P(A или B)= P(A) + P(B) - P(AB)$$



Пусть P(A) - вероятность выпадания на кубике больше 3 очков, то есть 4, 5 или 6.  
Какова вероятность такого события? Она равна сумме вероятностей выпадания 4, 5 и 6:
1/6 + 1/6 + 1/6 = 3/6 = 0.5
Тогда P(B) - вероятность выпадание больше 3 очков на 2м кубике, тоже равна 0.5
Подставим в вышеупомянутую формулу:
$$P(A+B)=P(A или B)= P(A) + P(B) - P(AB) = 0.5 + 0.5 - 0.5*0.5 = 0.75$$
